# Compute Average Node Length in Tree Structure

This notebook loads a JSONL file representing a tree structure, where each node is a text chunk and edges are defined by the `context_chunks` field. It computes the average node length and visualizes the distribution of node lengths.

```python
# Import Required Libraries
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
# Load JSONL Data
jsonl_path = Path('../data/raw/danish_dataset_poc/testing.jsonl')
nodes = []
with open(jsonl_path, 'r', encoding='utf-8') as f:
    for line in f:
        nodes.append(json.loads(line))
print(f"Loaded {len(nodes)} nodes.")
```

```python
# Parse Nodes and Edges
# Each node is a dict with 'chunk_id', 'chunk', and 'context_chunks' (edges)
nodes_df = pd.DataFrame(nodes)
nodes_df.head()
```

```python
# Compute Node Lengths
nodes_df['node_length'] = nodes_df['chunk'].apply(len)
nodes_df[['chunk_id', 'node_length']].head()
```

```python
# Calculate Average Node Length
average_length = nodes_df['node_length'].mean()
print(f"Average node length: {average_length:.2f} characters")
```

```python
# Display Results
plt.figure(figsize=(10, 5))
sns.histplot(nodes_df['node_length'], bins=30, kde=True)
plt.title('Distribution of Node Lengths')
plt.xlabel('Node Length (characters)')
plt.ylabel('Frequency')
plt.show()
```
